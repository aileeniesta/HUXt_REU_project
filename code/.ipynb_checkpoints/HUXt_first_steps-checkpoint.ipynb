```python
import numpy as np
import pandas as pd
import astropy.units as u
import matplotlib.pyplot as plt
import datetime
import os
import sys
sys.path.insert(0,"../HUXt/code")
sys.path.append('./orbitals_download/')
import importlib
import helpers
import huxt as H
import huxt_analysis as HA
import huxt_inputs as Hin
import sunpy
from sunpy.coordinates import sun
import httplib2
from scipy.interpolate import interp1d
from pandas import Series
from astropy.time import Time
import pickle

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "axes.labelsize": 14,
    "axes.titlesize": 14,
    "legend.fontsize": 11,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "axes.linewidth": 0.8,
    "lines.linewidth": 1.2
})
```

```python
# This is the wrapped function to run for diff. encounters.
def run_huxt_for_encounter(encounter_number):


    def get_cr_info(timestamp):
        cr1_start = Time('1853-11-09 12:00:00', format='iso', scale='utc')
        cr_length = 27.2753
        cr_num = int(((Time(timestamp) - cr1_start).to('day').value) / cr_length) + 1
        cr_start = cr1_start + (cr_num - 1) * cr_length * u.day
        return cr_num, cr_start.to_datetime()

    def compute_time_days(df, cr_start_dt):
        df['Times'] = pd.to_datetime(df['Times'], errors='coerce')
        return (df['Times'] - cr_start_dt).dt.total_seconds() / 86400

    def compute_adjusted_vr(df, model_velocity_set, model_distance):
        adjusted = []
        for i in range(len(df)):
            v0 = df.iloc[i]['Vr']
            r0 = df.iloc[i]['radius']
            r_idx = np.argmin(np.abs(model_distance - r0))
            v_col = model_velocity_set[:, r_idx]
            v_idx = np.argmin(np.abs(v_col - v0))
            v_profile = model_velocity_set[v_idx, :]
            r_30 = np.argmin(np.abs(model_distance - 30))
            adjusted.append(v_profile[r_30])
        return np.array(adjusted)

    def compute_rmse(a1, a2):
        a1, a2 = np.array(a1), np.array(a2)
        mask = ~np.isnan(a1) & ~np.isnan(a2)
        return np.sqrt(np.mean((a1[mask] - a2[mask])**2)) if np.sum(mask) > 0 else np.nan

    # --- Load data 
    df = helpers.get_prograde_df(encounter_number)
    df['Times'] = pd.to_datetime(df['time'], errors='coerce')
    df_full = pd.read_csv(f'../data/E{encounter_number:02d}.csv')
    df_full.columns = [col.replace('-', '_').replace('â€“', '_').strip() for col in df_full.columns]
    df_full['Times'] = pd.to_datetime(df_full['Times'], errors='coerce')

    # --- Carrington rotation and compute Time_days 
    cr, cr_start_dt = get_cr_info(df['Times'].iloc[0])
    df['Time_days'] = compute_time_days(df, cr_start_dt)
    df_full['Time_days'] = compute_time_days(df_full, cr_start_dt)

    # --- Load model profiles and adjust PSP Vr 
    models = pickle.load(open('../data/the_models_2025-05-01','rb'))
    model_distance = models[0][0].to_value(u.solRad)
    model_velocity_set = np.array([v.to_value(u.km/u.s) for v in models[0][2]])
    adjusted_vr = compute_adjusted_vr(df, model_velocity_set, model_distance)

    # --- Interpolate adjusted vr onto 1 deg grid and merge with MAS
    lon_uniform = np.arange(0, 360, 1)
    interp_adj = interp1d(df['longitude'], adjusted_vr, kind='linear', bounds_error=False, fill_value=np.nan)
    vr_downsampled = interp_adj(lon_uniform)
    v_mas = Hin.get_MAS_long_profile(cr, 0.0*u.deg)
    adjusted_vr_merged = np.where(np.isnan(vr_downsampled), v_mas.value, vr_downsampled) * v_mas.unit

    # --- Run HUXt 
    model = H.HUXt(v_boundary=adjusted_vr_merged, cr_num=cr, simtime=27*u.day, dt_scale=4, r_min=30*u.solRad)
    model.solve([])
    model_mas = H.HUXt(v_boundary=v_mas, cr_num=cr, simtime=27*u.day, dt_scale=4, r_min=30*u.solRad)
    model_mas.solve([])

    # --- Determine influence zone from longitude bounds 
    long_min, long_max = df['longitude'].min(), df['longitude'].max()
    time_start = (helpers.delta_long(30.665781*u.R_sun, r_inner=1*u.au, vsw=400.*u.km/u.s) + long_min*u.deg)/(360*u.deg) * 27.2753
    time_end   = (helpers.delta_long(30.648218*u.R_sun, r_inner=1*u.au, vsw=400.*u.km/u.s) + long_max*u.deg)/(360*u.deg) * 27.2753

    # --- Interpolate L1 and compute RMSE in grey zone 
    interp_L1 = interp1d(df_full['Time_days'], df_full['Vpr_L1'], kind='linear', bounds_error=False, fill_value="extrapolate")
    time_mas = np.linspace(df_full['Time_days'].min(), df_full['Time_days'].max(), len(v_mas))
    v_L1_interp = interp_L1(time_mas)
    mask = (time_mas >= time_start.value) & (time_mas <= time_end.value)
    rmse_adj = compute_rmse(adjusted_vr_merged[mask], v_L1_interp[mask])
    rmse_mas = compute_rmse(v_mas[mask], v_L1_interp[mask])

    # -- Plotting 
    fig1, ax1 = HA.plot_timeseries(model, 1.0*u.au, 0.0*u.deg, tag='v_merged')
    ax1.lines[0].set_color('red')
    ax1.lines[0].set_label(f'adjusted_vr (RMSE={rmse_adj:.1f})')

    fig2, ax2 = HA.plot_timeseries(model_mas, 1.0*u.au, 0.0*u.deg, tag='v_mas')
    ax1.plot(ax2.lines[0].get_xdata(), ax2.lines[0].get_ydata(), label=f'v_mas (RMSE={rmse_mas:.1f})', color='black')

    ax1.plot(df_full['Time_days'], df_full['Vpr_L1'], label='L1 Observed', color='blue')
    ax1.axvspan(time_start.value, time_end.value, color='silver', alpha=0.4, label='PSP Influence')
    ax1.axvline(time_start.value, color='black', linestyle='--')
    ax1.axvline(time_end.value, color='black', linestyle='--')

    ax1.set_xlabel("Time (days)")
    ax1.set_ylabel("Solar Wind Speed (km/s)")
    ax1.set_title(f"HUXt Prediction vs L1 Observation (ENC {encounter_number})")
    ax1.legend()
    ax1.grid(True)
    fig1.tight_layout()
    plt.show()

    return {
        'cr': cr,
        'long_min': long_min,
        'long_max': long_max,
        'time_start': time_start,
        'time_end': time_end,
        'rmse_adj': rmse_adj,
        'rmse_mas': rmse_mas
    }

```

```python
def get_v_merged(encounter_number, v_mas):
    """
    Returns the HUXt-ready v_merged boundary and auto-detects Carrington rotation number
    based on the encounter time.
    """
    # Step 1: Load PSP prograde data
    df = helpers.get_prograde_df(encounter_number)

    # Step 2: Infer Carrington Rotation number
    def get_cr_number_from_time(timestamp):
        cr1_start = Time('1853-11-09 12:00:00', format='iso', scale='utc')
        cr_length = 27.2753  # days
        t = Time(timestamp)
        return int(((t - cr1_start).to('day').value) / cr_length) + 1

    cr_number = get_cr_number_from_time(df['time'].iloc[0])

    # Step 3: Interpolate Vr onto 1-degree longitude grid
    lon_uniform = np.arange(0, 360, 1)
    interp_func = interp1d(df['longitude'], df['Vr'], kind='linear',
                           bounds_error=False, fill_value=np.nan)
    vr_downsampled = interp_func(lon_uniform)

    # Step 4: Merge with v_mas to fill gaps
    v_merged_values = np.where(np.isnan(vr_downsampled), v_mas.value, vr_downsampled)
    v_merged = v_merged_values * v_mas.unit

    # Step 5: Return v_merged and Carrington rotation number
    return v_merged, cr_number

```

```python
enc = 22
df = pd.read_csv(f'../data/E{enc:02d}.csv')
df.columns = [col.replace('-', '_').replace('â€“', '_').strip() for col in df.columns]
df['Times'] = pd.to_datetime(df['Times'], errors='coerce')

cr_start = pd.to_datetime('2024-12-10')
cr_end = pd.to_datetime('2025-01-06')
df = df[(df['Times'] >= cr_start) & (df['Times'] <= cr_end)].copy()

# --- Time in days relative to CR start ---
df['Time_days'] = (df['Times'] - cr_start).dt.total_seconds() / 86400

# --- Run HUXt with v_merged ---
v_merged, cr = get_v_merged(encounter_number=22, v_mas=v_mas)
model = H.HUXt(v_boundary=v_merged, cr_num=cr, simtime=27*u.day, dt_scale=4, r_min=30*u.solRad)
model.solve([])

model_mas = H.HUXt(v_boundary=v_mas, cr_num=cr, simtime=27*u.day, dt_scale=4, r_min=30*u.solRad)
model_mas.solve([])

# --- Compute RMSE before plotting ---
from scipy.interpolate import interp1d

# Interpolate L1 data to model timescale
interp_L1 = interp1d(df['Time_days'], df['Vpr_L1'], kind='linear', bounds_error=False, fill_value="extrapolate")
time_mas = np.linspace(df['Time_days'].min(), df['Time_days'].max(), len(v_mas))
v_L1_interp = interp_L1(time_mas)

# RMSE function
def compute_rmse_ignore_nan(a1, a2):
    a1 = np.array(a1)
    a2 = np.array(a2)
    mask = ~np.isnan(a1) & ~np.isnan(a2)
    if np.sum(mask) == 0:
        return np.nan
    return np.sqrt(np.mean((a1[mask] - a2[mask]) ** 2))

# Compute RMSEs
rmse_mas_vs_l1 = compute_rmse_ignore_nan(v_mas, v_L1_interp)
rmse_merged_vs_l1 = compute_rmse_ignore_nan(v_merged, v_L1_interp)
rmse_mas_label = f'V_mas (RMSE={rmse_mas_vs_l1:.1f})'
rmse_merged_label = f'v_merged (RMSE={rmse_merged_vs_l1:.1f})'

# --- Plot merged model ---
fig1, ax1 = HA.plot_timeseries(model, r, lon=0.0*u.deg, tag='v_merged')

# Recolor and relabel v_merged
ax1.lines[0].set_color('red')
ax1.lines[0].set_label(rmse_merged_label)

# --- Plot MAS model ---
fig2, ax2 = HA.plot_timeseries(model_mas, r, lon=0.0*u.deg, tag='v_mas')
mas_line = ax2.lines[0]

# Add MAS line to merged plot
ax1.plot(mas_line.get_xdata(), mas_line.get_ydata(), label=rmse_mas_label, color='black', linestyle='-')

# --- Plot L1 observations ---
ax1.plot(df['Time_days'], df['Vpr_L1'], label='L1 Observed', color='blue', linewidth=1)

# --- PSP influence zone ---
ax1.axvspan(time_start.value, time_end.value, color='silver', alpha=0.4, label='PSP Influence')
ax1.axvline(time_start.value, color='black', linestyle='-', linewidth=1)
ax1.axvline(time_end.value, color='black', linestyle='-', linewidth=1)

# --- Axis formatting ---
ax1.set_xlabel("Time (days)")
ax1.set_ylabel("Solar Wind Speed (km/s)")
ax1.set_title(f"HUXt Prediction vs L1 Observation (ENC {enc})")
ax1.legend()
ax1.grid(True)
fig1.tight_layout()
plt.show()
```

```python
def compute_rmse_ignore_nan(a1, a2):
    a1 = np.asarray(a1)
    a2 = np.asarray(a2)
    mask = ~np.isnan(a1) & ~np.isnan(a2)
    if np.sum(mask) == 0:
        return np.nan
    return np.sqrt(np.mean((a1[mask] - a2[mask]) ** 2))
```

```python
rmse_mas_vs_l1 = compute_rmse_ignore_nan(v_mas, v_L1_interp)
rmse_merged_vs_l1 = compute_rmse_ignore_nan(v_merged, v_L1_interp)

print("RMSE (v_mas vs v_L1):", rmse_mas_vs_l1)
print("RMSE (v_merged vs v_L1):", rmse_merged_vs_l1)

# range_L1 = np.nanmax(v_L1_interp) - np.nanmin(v_L1_interp)
# nrmse_mas = rmse_mas_vs_l1 / range_L1
# nrmse_merged = rmse_merged_vs_l1 / range_L1

# print("Normalized RMSE (v_mas):", nrmse_mas)
# print("Normalized RMSE (v_merged):", nrmse_merged)
```
